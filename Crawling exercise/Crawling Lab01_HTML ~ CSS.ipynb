크롤링 핵심 코드 패턴으로 이해하기

1. import requests
   from bs4 import BeautifulSoup >>  requests, bs4 라이브러리 불러오기
2. res = requests.get() >> 크롤링 할 웹사이트 주소
3. soup = BeautifulSoup(res.content, 'html.parser') >> 위에 res 변수에 저장한 크롤링 대상 사이트를 parsing 하여 웹사이트 정보 취득 및 분류
4. mydata = soup.find() >> parsing을 통해 정리한 웹사이트의 정보 중 크롤링 할 내용을 선정하고 크롤링 (html 기본 구조를 알 필요가 있음)
5. print(mydata.get_text()) >> mydata에 저장 된 크롤링 정보를 get_text 함수를 이용해 추출

결론적으로 위의 구조에서 
{1. 크롤링 할 웹사이트 주소(requests.get 이용)} {2. (soup.find)를 이용해 크롤링 할 내용} 
두 가지만 바꿔주면 웬만한 것들은 크롤링이 가능.


```python
import requests
from bs4 import BeautifulSoup
res = requests.get('http://www.kwnews.co.kr/page/view/2023042516031550861')
soup = BeautifulSoup(res.content, 'html.parser')
mydata = soup.find('title')
print(mydata.get_text())
        
```

패턴으로 실습하며 익히기: HTML 이해를 바탕으로 크롤링하기

```python
from bs4 import BeautifulSoup
html ="""<html>
        <body>
            <h1 id = 'title'>[1]크롤링이란?</h1>
             <p class = 'cssstyle'>웹페이지에서 필요한 데이터를 추출하는 것</p>
             <p id='body' align='center'>파이썬을 중심으로 다양한 웹크롤링 기술 발달</p>
        </body>
"""

soup = BeautifulSoup(html, "html.parser")
#태그로 검색 방법
data = soup.find('h1')
print(data)
print(data.string)
print(data.get_text())
```

```python
#크롤링을 할 때 속성을 정하는 방법은 여러가지가 있음
data = soup.find('p', class_='cssstyle')
data = soup.find('p', 'cssstyle')
data = soup.find('p', attrs = {'id':'body', 'align':'center'})
data = soup.find(id='body')
data = soup.find_all('p')
```

```python
#find_all + 반복문을 이용한 크롤링
from bs4 import BeautifulSoup
html ="""<html>
        <body>
            <h1 id = 'title'>[1]크롤링이란?</h1>
             <p class = 'cssstyle'>웹페이지에서 필요한 데이터를 추출하는 것</p>
             <p id='body' align='center'>파이썬을 중심으로 다양한 웹크롤링 기술 발달</p>
        </body>
"""

soup = BeautifulSoup(html, "html.parser")
#태그로 검색 방법
data = soup.find_all('p')

#find_all을 이용해 모든 p 데이터를 크롤링하였기에 출력하려면 반복문을 이용해야함
for text in data:
    print(text.get_text())
```

```python
#G마켓 베스트 아이템 크롤링
import requests
from bs4 import BeautifulSoup
res = requests.get('https://www.gmarket.co.kr/n/best')
soup = BeautifulSoup(res.content, 'html.parser')
mydata = soup.find_all(class_='itemname')
for bestitem in mydata:
    print(bestitem.get_text())
        
```
